```python
import requests
import datetime
import pandas as pd
```

```python
# Get daily trade data of BTC
url = 'https://min-api.cryptocompare.com/data/v2/histoday?fsym=BTC&tsym=USD&allData=true'
page = requests.get(url)
data = page.json()['Data']
df = pd.DataFrame(data)
```

```python
df
```

```python
# Extract data
keys = df['Data'][0].keys()
df_trade_data = pd.DataFrame()
for key in keys:
    df_trade_data[key] = df['Data'].apply(lambda x: x.get(key))
# Drop useless columns
df_trade_data.drop(columns=['conversionType', 'conversionSymbol'], inplace=True)
df_trade_data
```

```python
# Change Unix timestamp into ISO 8601
def timestampToISO(x):
    dt = datetime.datetime.utcfromtimestamp(x)
    dt = dt.replace(microsecond=0)
    iso_string = dt.isoformat() + '+00:00'
    return iso_string
df_trade_data['time'] = df_trade_data['time'].apply(lambda x: timestampToISO(x))
df_trade_data
```

```python
# Save as csv
df_trade_data.to_csv('marketdata.csv', encoding='utf-8', index=False)
```
